```python
# Clean the 'categories' column - remove brackets and quotes
import ast
import numpy as np

def clean_category(x):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return None
    # If it is a string that looks like a list
    if isinstance(x, str):
        try:
            lst = ast.literal_eval(x)
            if isinstance(lst, list) and len(lst) > 0:
                return lst[0].strip()  # take first element
        except (ValueError, SyntaxError):
            return x.strip()  # fallback: just clean string
    # If it is already a list
    elif isinstance(x, list) and len(x) > 0:
        return x[0].strip()
    return None

df4['categories'] = df4['categories'].apply(clean_category)
```

```python
missing_titles = df4[df4['categories'].isnull() | (df4['categories'] == '')]['Title'].unique()
print("Titles with missing categories:", len(missing_titles))
print(missing_titles[:10])  # show first 10
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm

# --- Assume df3 is your dataframe ---
# Columns: ['Title', 'categories']

# Step 1: Separate known and missing
known = df4.dropna(subset=['categories']).copy()
missing = df4[df4['categories'].isna()].copy()

print("Known categories:", len(known))
print("Missing categories:", len(missing))

# Step 2: Fit TF-IDF on all titles
vectorizer = TfidfVectorizer(stop_words='english', max_features=20000)
vectorizer.fit(df4['Title'].astype(str))

known_vecs = vectorizer.transform(known['Title'].astype(str))
missing_vecs = vectorizer.transform(missing['Title'].astype(str))

# Step 3: Compute similarity in batches (to handle large data)
batch_size = 1000
inferred_categories = []

for start in tqdm(range(0, missing_vecs.shape[0], batch_size)):
    end = start + batch_size
    batch_vecs = missing_vecs[start:end]

    # Compute cosine similarity with known titles
    sim = cosine_similarity(batch_vecs, known_vecs)

    # Find best match index for each missing title
    best_match_idx = sim.argmax(axis=1)
    best_categories = known.iloc[best_match_idx]['categories'].values

    inferred_categories.extend(best_categories)

# Step 4: Assign inferred categories back
missing['categories'] = inferred_categories

# Step 5: Merge back into main dataframe
df4.loc[df4['categories'].isna(), 'categories'] = missing['categories']

print("✅ Missing categories filled using title similarity.")
```

```python
# Define a mapping function for sentiment analysis
def map_sentiment(score):
    if score in [1, 2]:
        return 'negative'
    elif score == 3:
        return 'neutral'
    elif score in [4, 5]:
        return 'positive'
    else:
        return None
```

```python
import nltk

# Step 2: Download required datasets
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import re #regular expressions (for pattern matching, e.g., removing numbers or symbols)
import string #tring operations (e.g., punctuation removal)
from nltk.corpus import stopwords #common words like “the”, “is”, “and” that don’t add meaning
from nltk.stem import WordNetLemmatizer #reduces words to their base form (“running” → “run”)
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if pd.isna(text):
        return ''  # handle missing reviews

    # Lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove numbers and non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenize and remove stopwords
    tokens = [word for word in text.split() if word not in stop_words]

    # Lemmatize words
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens
```
